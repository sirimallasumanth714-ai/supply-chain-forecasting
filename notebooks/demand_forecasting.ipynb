```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Load your sales data
df = pd.read_csv("../data/sales_data.csv")
df['date'] = pd.to_datetime(df['date'])
df = df.rename(columns={"date": "ds", "units_sold": "y"})

# Visualize historical demand
plt.figure(figsize=(10,5))
plt.plot(df['ds'], df['y'], marker='o')
plt.title("Historical Demand")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()

# Train the Prophet forecasting model
model = Prophet()
model.fit(df)

# Forecast the next 30 days
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Plot the forecast
fig1 = model.plot(forecast)
plt.show()

# Save plot to data folder for README
fig1.savefig("../data/forecast_plot.png")


# Save forecast results to CSV
forecast[['ds','yhat','yhat_lower','yhat_upper']].to_csv("../data/forecast_results.csv", index=False)
print("Forecast saved to ../data/forecast_results.csv")
```
